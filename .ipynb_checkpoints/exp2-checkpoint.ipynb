```python
import seaborn as sns
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
bank_dataset = pd.read_csv('./exp1-new.csv')
print(bank_dataset.head())
                           
```

```python
plt.figure(figsize = (15, 8), dpi = 100) 
color_palette = sns.color_palette("Accent_r") 
sns.set_palette(color_palette)
sns.countplot(x = "education", data = bank_dataset) 
```

```python
numerical_columns = bank_dataset.select_dtypes(include=['int64', 'float64']).columns
bank_dataset_numeric = bank_dataset[numerical_columns]

plt.figure(figsize=(10, 6), dpi=100)
color = sns.color_palette("magma")
sns.heatmap(bank_dataset_numeric.corr(), vmax=0.9, annot=True, cmap=color)
```

```python
plt.figure(figsize=(15,7), dpi=100)
color_palette = sns.color_palette("magma")
sns.set_palette(color_palette)
sns.barplot(x="education", y="education", data=bank_dataset)
```

```python
education_counts = bank_dataset['education'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(education_counts, labels=education_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Proportion of Education Levels')
plt.axis('equal')  
plt.show()
```

```python
contact_freq_by_month = bank_dataset['month'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
plt.plot(contact_freq_by_month.index, contact_freq_by_month.values, marker='o', linestyle='-')
plt.title('Contact Frequency Over Months')
plt.xlabel('Month')
plt.ylabel('Frequency')
plt.grid(True)
plt.xticks(rotation=45)  
plt.tight_layout()
plt.show()
```
