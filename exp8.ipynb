```python
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score

bank_dataset = pd.read_csv('exp5.csv')
bank_dataset.head()
```

```python
bank_dataset_copy = bank_dataset.copy()

x = bank_dataset_copy[['age_normalized' , 'duration' , 'campaign' , 'pdays' , 'previous', 'loan','day_of_week', 'housing',  'marital', 'education', 'job', 'contact', 'month', 'poutcome']]
y = bank_dataset_copy['y']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
#Classification model
model = LogisticRegression(max_iter=5000) 
model.fit(X_train, y_train) 
y_pred = model.predict(X_test) 
```

```python
from sklearn.metrics import confusion_matrix 
cm = confusion_matrix(y_test, y_pred)
print(accuracy_score(y_test, model.predict(X_test)))
```
