```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
bank_dataset = pd.read_csv('exp5.csv')
bank_dataset.head()
```

```python
x = bank_dataset.iloc[:,0:11] # 1st for rows and second for columns
x
```

```python
kmeans = KMeans(3) # Creating three clusters
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = bank_dataset.copy()
data_with_clusters['Clusters'] = identified_clusters 
plt.scatter(data_with_clusters['age'],data_with_clusters['duration'],c=data_with_clusters['Clusters'],cmap='rainbow')
```
